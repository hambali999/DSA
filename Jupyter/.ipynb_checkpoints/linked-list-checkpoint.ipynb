```python
# step 4 - insert at end

class Node:
    def __init__(self, data=None, next=None): # data itself, next acts as a pointer to point to the next Node
        self.data = data # the data itself
        self.next = next # pointer to point to the next node
        
        print(self.data)
        print(self.next)
        print("----------------------")
        
class LinkedList: #creation of the linked_list
    def __init__(self):
        self.head = None #when creating an empty linked_list, head == None, cos no nodes added yet
     
    # inserting for the first time, data = data, self.head/next = None
    # inserting for the second time, data = data, self.head/next = will the previous node
    def insert_at_beginning(self, data): # inserting a node at the beginning with the data
        node = Node(data, self.head) # self.head == the previous node self.head, acts as the next pointer since we adding in the beginning.
        self.head = node #now when adding new node in the beginning, declare it as the self.head so that the next node have the pointer to this node self.haed
        
    def print(self):
        if self.head == None: #if linked-list is empty
            print("Linked list is empty!")
        else:
            llstr = ''
            itr = self.head
            # while itr:
            while itr is not None:
                llstr += (str(itr.data) + "-->" )
                itr = itr.next
            print(llstr)
    
    def insert_at_end(self, data):
        if self.head is None: #when linked list is blank
            self.head = Node(data, None) #since we're inserting a Node at the end, the next element would be null if its an empty linked list 
            return
        # if linked-list is not blank, iterate through the linked-list and at the end insert the node
        itr = self.head
        while itr.next: #if itr.next has some values, means its not at the end
            itr = itr.next
        # when itr.next becomes null, it means we're at the last element
        itr.next = Node(data, None)
            
            
            
    
if __name__ == '__main__':
    ll = LinkedList()
    ll.insert_at_beginning(5)
    ll.insert_at_beginning(89)
    ll.insert_at_beginning(26)
    ll.insert_at_end(79)
    ll.insert_at_end(128)

    ll.print()

        
```

```python
# step 10 - remove by value

class Node:
    def __init__(self, data=None, next=None): # data itself, next acts as a pointer to point to the next Node
        self.data = data # the data itself
        self.next = next # pointer to point to the next node
        
        #print(self.data)
        #print(self.next)
        #print("----------------------")
        
class LinkedList: #creation of the linked_list
    def __init__(self):
        self.head = None #when creating an empty linked_list, head == None, cos no nodes added yet
     
    # inserting for the first time, data = data, self.head/next = None
    # inserting for the second time, data = data, self.head/next = will the previous node
    def insert_at_beginning(self, data): # inserting a node at the beginning with the data
        node = Node(data, self.head) # self.head == the previous node self.head, acts as the next pointer since we adding in the beginning.
        self.head = node #now when adding new node in the beginning, declare it as the self.head so that the next node have the pointer to this node self.haed
        
    def print(self):
        if self.head == None: #if linked-list is empty
            print("Linked list is empty!")
        else:
            llstr = ''
            itr = self.head
            # while itr:
            while itr is not None:
                llstr += (str(itr.data) + "-->" )
                itr = itr.next
            print(llstr)
    
    def insert_at_end(self, data):
        if self.head is None: #when linked list is blank
            self.head = Node(data, None) #since we're inserting a Node at the end, the next element would be null if its an empty linked list 
            return
        # if linked-list is not blank, iterate through the linked-list and at the end insert the node
        itr = self.head
        while itr.next: #if itr.next has some values, means its not at the end
            itr = itr.next
        # when itr.next becomes null, it means we're at the last element
        itr.next = Node(data, None)
        
    def insert_values(self, data_list): #to create a fresh new linked-list
        self.head = None
        for data in data_list:
            self.insert_at_end(data)
            
    def get_length(self): #to get the length of the linked-list
        count = 0
        itr = self.head
        while itr:
            count+=1
            itr = itr.next
        return count
    
    def remove_at(self, index):
        # check if index is valid
        if index < 0 or index >= self.get_length():
            raise Exception("Not a valid index.")
        
        if index == 0: #garbage collection - python will help us to free the memory automatically
            self.head = self.head.next #now the new self.head is pointing to the next element self.head since we are removing at the beginning index.
            return 
        
        count = 0
        #loop each element in a linked-list
        itr = self.head
        while itr:
            #stop at the element index we want to remove.
            #in that element, we have to modify the links/pointers that we have.
            #modifying the links/pointers automatically remove the element as we dont have any corelation pointers anymore.
            if count == index - 1: #we are at the previous element
                itr.next = itr.next.next #point to the next's element next.
                break
                
            itr = itr.next
            count += 1
            
    def insert_at(self, index, data):
        # check if index is valid
        if index < 0 or index >= self.get_length():
            raise Exception("Not a valid index.")
            
        # insert at beginning
        if index == 0:
            self.insert_at_beginning(data)
            return
        
        # insert at an index
        count = 0
        itr = self.head
        while itr:
            #if we are inserting an element in a linked-list, we want to modify the next pointer of the previous element,
            #
            if count == index - 1: #we are at the previous element
                #previous element next will be this node next. because we are inserting it in between
                node = Node(data, itr.next) #the new node next pointer will be the the previous node's next
                itr.next = node #now the previous element next's, will be the node we are inserting hence this line.
                break
            itr = itr.next
            count+=1
            
    def insert_after_value(self, data_after, data_to_insert):
        
        if self.head is None:
            return

        if self.head.data==data_after:
            self.head.next = Node(data_to_insert,self.head.next)
            return
        
        #insert at a given value 'data_after'
        count = 0
        itr = self.head #self.head = node
        while itr: #iterating through the nodes
            if data_after == itr.data: #if there is a matching node data,
                node = Node(data_to_insert, itr.next) #create the node, its next will point to the previus elemnet next's
                itr.next = node
                break
            #print(itr.data)
            itr = itr.next
            count+=1
            
    def remove_by_value(self, data):
        
        if self.head is None:
            return

        if self.head.data == data:
            self.head = self.head.next
            return
        
        #remove the node given from the parameter 'data'
        #lets iterate through the nodes
        count = 0
        itr = self.head
        #iterating through the nodes
        while itr.next:
            if itr.next.data == data:
                itr.next = itr.next.next
                break
            itr = itr.next
            
    
            
    
if __name__ == '__main__':
    ll = LinkedList()
    ll.insert_values(["Apple", "Banana", "Watermelon", "Pineapple"])
#     ll.print()
#     ll.remove_at(2)
#     ll.print()
#     ll.insert_at(1, "Orange")
#     ll.print()
#     print(ll.get_length())
    ll.insert_after_value("Banana", "KIKI")
    ll.print()
    ll.remove_by_value("Banana")
    ll.print()

        
```
